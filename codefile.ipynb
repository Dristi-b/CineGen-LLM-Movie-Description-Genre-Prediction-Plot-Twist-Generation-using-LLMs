***Task 1 : Use LLM to generate a new creative description based on title and description from the dataset and return the result in a json format containg   {movie_title,orginal_description,generated_description,genres,cast}***

```python
# Load the dataset to feed titles into the LLM for tasks
print("Dataset:")
df = pd.read_csv("movie_dataset.csv")

# to avoid inconsistencies in column naming
if 'title' not in df.columns:
    df.rename(columns={'movie_title': 'title'}, inplace=True)

df = df.head(20)  # setting the limitations

# Initializing LLM pipeline using Hugging Face transformers
generator = pipeline("text-generation", model="microsoft/phi-2", device_map="auto")
#Initializes the text generation pipeline with the 'microsoft/phi-2' model.
#device_map="auto" ensures it uses GPU if available, else CPU automatically.

# function for creative description of the movies
def generate_description(title, original_description):
    prompt = (
        f"Generate a creative and engaging movie description.\n"
        f"Title: {title}\n"
        f"Original Description: {original_description}\n"
        f"Creative Description:"
    )
    output = generator(prompt, max_new_tokens=200, do_sample=True, temperature=0.9)[0]["generated_text"] #calling the generator with the prompt given
    generated = output.split("Creative Description:")[-1].strip() #extracting only the generated description portion

    #removing unwanted text in the code below
    if "Title:" in generated:
        generated = generated.split("Title:")[0].strip()
    if "Real-World Use Case" in generated or "Exercise" in generated:
        generated = generated.split("\n")[0].strip()
    return generated

# Generating Task 1
results_task1 = []
print("Generating creative descriptions :")
for idx, row in tqdm(df.iterrows(), total=len(df)): # progress bar visualization in the output while iterating through 20 movies
    title = row["title"]
    description = row["description"]
    try:
        generated_desc = generate_description(title, description)
    except Exception as e:
        print(f"Error on {title}: {e}")
        generated_desc = ""
    results_task1.append({
        "movie_title": title,
        "original_description": description,
        "generated_description": generated_desc,
        "genres": row.get("genres", ""),
        "cast": row.get("cast", "")
    })
#Saving the generated descriptions to a JSON file
with open("task1.json", "w") as f:
    json.dump(results_task1, f, indent=4)
print("Task 1 completed and saved as task1.json.")
```

```python
from transformers import pipeline  # imported to load Hugging Face zero-shot classification model
import json
import pandas as pd
from tqdm import tqdm

# Load movie dataset JSON directly
with open("movie_dataset.json", "r") as f:
    data = json.load(f)

df_task2 = pd.DataFrame(data)

# Initialized zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# genre categories from my dataset
possible_genres = [
    'Action', 'Adventure', 'Animation', 'Biography', 'Comedy',
    'Crime', 'Drama', 'Family', 'Fantasy', 'History',
    'Horror', 'Music', 'Musical', 'Mystery', 'Romance',
    'Sci-Fi', 'Sport', 'Thriller', 'War', 'Western'
]

# Clean descriptions if needed
def clean_description(text):
    return text.strip().replace("\n", " ")[:512]

results_task2 = []  # list to store task 2
correct_count = 0

print("Predicting genres using zero-shot classification :")

for idx, row in tqdm(df_task2.iterrows(), total=len(df_task2)):
    description = row["description"]
    actual_genres = [g.strip().lower() for g in row["genres"]]

    try:
        result = classifier(description, possible_genres, multi_label=True)
        # Take top 3 genres with highest score, no threshold filtering
        predicted_genres_list = result['labels'][:3]
        predicted_genres = ", ".join(predicted_genres_list)
    except Exception as e:
        print(f"Error at index {idx}: {e}")
        predicted_genres = ""
        predicted_genres_list = []

    # Checking if any predicted genre matches the actual genre (case-insensitive)
    match_found = any(pred.lower() in actual_genres for pred in predicted_genres_list)
    if match_found:
        correct_count += 1

    results_task2.append({
        "movie_title": row["movie_title"],
        "description": description,
        "actual_genres": actual_genres,
        "predicted_genres": predicted_genres,
        "match": match_found
    })

accuracy = correct_count / len(df_task2)
print(f"Genre Prediction Accuracy: {accuracy * 100:.2f}%")

# Saving output
with open("task2.json", "w") as f:
    json.dump(results_task2, f, indent=4)

print("Task 2 completed and saved as task2.json.")
```

```python
import matplotlib.pyplot as plt
import json

# Loading Task 2 JSON file
with open("task2.json") as f:
    genre_data = json.load(f)

correct = sum(1 for entry in genre_data if entry["match"])
incorrect = len(genre_data) - correct

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie([correct, incorrect], labels=["Correct Matches", "Incorrect Matches"], autopct='%1.1f%%', colors=["#4CAF50", "#FF5722"])
plt.title("Genre Prediction Accuracy (Task 2)")
plt.show()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
import pandas as pd
df = pd.read_csv("movie_dataset.csv")
df.columns = df.columns.str.strip().str.lower()  # Normalized the column names
if 'movie_title' not in df.columns or 'description' not in df.columns:
    raise ValueError("CSV must contain 'movie_title' and 'description' columns.")
```

```python
# loading the phi-2 model
import torch
model_name = "microsoft/phi-2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")
model.eval()

```

```python
from tqdm import tqdm
# Set the device for PyTorch (use GPU if available, otherwise fall back to CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Function to generate a plot twist for a given movie title and description.
def generate_plot_twist(title, description):
   # Now we will Construct the prompt for the language model to act like a creative screenwriter
    prompt = f"""Imagine you're a creative screenwriter. Given the movie description below, write a surprising plot twist or alternate ending in 1-3 sentences.

Movie Title: {title}
Description: {description}

Plot Twist:"""

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    # Generate output from the model using sampling for creativity
    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.9,
        top_p=0.95,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    # Decode the generated token IDs to actual text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the part after "Plot Twist:" to avoid repeating the whole prompt
    if "Plot Twist:" in generated_text:
        return generated_text.split("Plot Twist:")[-1].strip()
    else:
        return generated_text.strip()

# Now we will create an empty list to store the results
results = []

# Loop through each row in the DataFrame with a progress bar
for _, row in tqdm(df.iterrows(), total=len(df)):
    title = row['movie_title']
    description = row['description']
    twist = generate_plot_twist(title, description)
    results.append({
        "movie_title": title,
        "original_description": description,
        "plot_twist": twist
    })
```

```python
import json
# saving as a json file
with open("task3.json", "w") as f:
    json.dump(results, f, indent=4)

print(" Task 3 completed and saved as task3.json")
```

```python
import json
import pandas as pd
import os

#  to fix Unicode and Encoding in Text
def fix_text(text):
    if isinstance(text, str):
        return (
            text.encode('latin1', errors='ignore').decode('utf-8', errors='ignore')
            .replace('\u2019', "'")
            .replace('\u2018', "'")
            .replace('\u201c', '"')
            .replace('\u201d', '"')
            .replace('\u2013', '-')
            .replace('\u2026', '...')
            .replace('\u00a0', ' ')
            .replace('â€”', '—')
            .replace('â€™', "'")
            .replace('â€œ', '"')
            .replace('â€\x9d', '"')
            .replace('â€“', '-')
            .replace('â€¦', '...')
        )
    elif isinstance(text, list):
        return [fix_text(item) for item in text]
    elif isinstance(text, dict):
        return {key: fix_text(value) for key, value in text.items()}
    else:
        return text

#  Fix All JSON Files
json_files = ["task1.json", "task2.json", "task3.json", "movie_dataset.json"]

for file in json_files:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    cleaned = fix_text(data)

    with open(file, 'w', encoding='utf-8') as f:
        json.dump(cleaned, f, indent=2, ensure_ascii=False)

print(" JSON files cleaned successfully.")

#  cleaning CSV File
csv_file = "movie_dataset.csv"

df = pd.read_csv(csv_file, encoding='latin1', on_bad_lines='skip')

# Fix all text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(fix_text)

df.to_csv(csv_file, index=False, encoding='utf-8')
print("CSV file cleaned successfully.")
```
